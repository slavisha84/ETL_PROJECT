```python
# import dependencies
import pandas as pd
import numpy as np
import requests
import json
from pandas.io.json import json_normalize
```

```python
# Retrive curency exhcange using API for EUR to US conversion of the Value and Wage columns
url = "https://api.darksky.net/forecast/fb0e09b8e9092cef1008bfe2eef00118/33.972386,-84.231986"
response = requests.get(url)
data = response.json()
complete = json_normalize(data['hourly']['data'])
```

```python
complete.head()
```

```python
# Extracting the subset of values of interest
df = complete.loc[:,["time","temperature","pressure","humidity"]]

df.head()
```

```python
# Saving into excel file
df.to_csv("FORCASTING.csv", sep=',', encoding='utf-8')
```

```python
import os
import pandas as pd
```

```python
# Load in file
time_file = "Hourly_Forecast.csv"
```

```python
time_file_df = pd.read_csv(time_file)
time_file_df.head()
```

```python
pd.to_datetime(time_file_df['time'], unit='s')
```

```python
time_file_df['clean time'] = pd.to_datetime(time_file_df['time'], unit='s')
```

```python
time_file_df.head()
```

```python
# Create new table
new_time = time_file_df[["clean time", "temperature", "pressure","humidity"]]
new_time.head()
```

```python
# Convert temperature into celsius
new_time["Temp_c"] = round((new_time["temperature"]-32) * 5/9, 2)

forecast = new_time[["clean time", "Temp_c", "pressure","humidity"]]
forecast.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///Sensors.db', echo=False)
```

```python
# create sql table from panda df
new_time.to_sql('FORECAST', con=engine)
```
